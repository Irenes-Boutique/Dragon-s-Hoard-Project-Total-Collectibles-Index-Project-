```python
import pandas as pd
import numpy as np
```

```python
lego = pd.read_csv(r"C:\Users\varia\OneDrive\Dragon Hoard's Index\Datasets\LEGO BE100 Index.csv")
lego = lego.dropna()
```

```python
lego.head(10)
```

```python
name_list = lego['itemname'].drop_duplicates()
id_list = lego['itemid'].drop_duplicates()
print(f'{len(name_list)}, {len(id_list)}')
```

```python
print(type(lego.iloc[0,3]))
```

```python
lego['itemyear'] = lego['itemyear'].astype(int).astype(str)
lego['itemmonth'] = (lego['itemmonth']+1).astype(int).astype(str)
lego['itemday'] = lego['itemday'].astype(int).astype(str)

lego['datetime'] = lego['itemyear'] + '-' + lego['itemmonth'] + '-' + lego['itemday']
lego['datetime'] = pd.to_datetime(lego['datetime'])
```

```python
lego['price_avg'] = (lego['itemprice_lower'] + lego['itemprice_upper'])/2
```

```python
df = pd.DataFrame()

for item in id_list:
    # Get temp df with only data of specific id
    temp_df = lego[lego['itemid'] == str(item)].copy()
    price = temp_df.groupby('datetime')['price_avg'].mean().sort_index()
    df = df.join(price.rename(str(item)), how = 'outer')
```

```python
df = df.round(2)
df.tail()
```

```python
df.to_csv('BE100 Index.csv')
```
